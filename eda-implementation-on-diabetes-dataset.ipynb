```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#to load the dataset
df = pd.read_csv('/kaggle/input/eda-proj/Dataset.csv')

# to Print the summary
for column in df.columns:  
        min_val = df[column].min()
        max_val = df[column].max()
        mean_val = df[column].mean()
        std_val = df[column].std()
        missing_val = 1 if df[column].isnull().sum() > 0 else 0 

        print(f"Column: {column}")
        print(f"Min: {min_val}")
        print(f"Max: {max_val}")
        print(f"Mean: {mean_val}")
        print(f"Standard Deviation: {std_val}")
        print(f"Missing Values: {missing_val}")
        print("-" * 30)
```

```python
# Plot Pregnancies distribution 
plt.figure(figsize=(8,6))
sns.histplot(df['Pregnancies'])
plt.title('Distribution of Pregnancies')
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(df['Pregnancies'])
plt.title('Boxplot of Pregnancies')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(df['Pregnancies'])
plt.title('Scatter plot of Pregnancies')
plt.show()
```
